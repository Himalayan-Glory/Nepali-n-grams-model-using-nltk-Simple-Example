```python
tokens=text.split()
```

```python
print(tokens)
```

```python
#Do some text cleaning like punctuations and stopwords removing
```

```python
import string
punctuations= list(string.punctuation)
print(punctuations)
```

```python
punctuations.append('।')
```

```python
print(punctuations)
```

```python
cleaned_tokens=[w for w in tokens if w not in punctuations]
```

```python
print(cleaned_tokens)
```

```python
text_unigram = ngrams(cleaned_tokens, 1)
Counter(text_unigram).most_common(20)
```

```python
# Checking the most commong bigrams
text_bigrams = ngrams(cleaned_tokens, 2)
Counter(text_bigrams).most_common(20)
```

```python
# Checking the most commong trigrams
text_trigrams = ngrams(cleaned_tokens, 3)

Counter(text_trigrams).most_common(10)
```
